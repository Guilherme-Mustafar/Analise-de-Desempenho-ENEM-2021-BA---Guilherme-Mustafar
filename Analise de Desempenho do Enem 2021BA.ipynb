## Etapa 1 - Ingest√£o e Auditoria Inicial dos Dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dados_enem_2021_BA.csv')
pd.set_option('display.max_columns', 100)
df.head()
```

```python
#Cria√ß√£o de uma coluna que calcula a media de todos as notas
#Vers√£o preliminar para auditoria de qualidade
df['MediaNotas'] = df[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT', 'NU_NOTA_REDACAO']].mean(axis=1)

df['qtd_notas_validas'] = df[['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 
                              'NU_NOTA_MT', 'NU_NOTA_REDACAO']].notna().sum(axis=1)

df.loc[df['qtd_notas_validas'] < 3, 'MediaNotas'] = np.nan
```

```python
df['qtd_notas_validas'].value_counts()
```

```python
df['MediaNotas']
```

```python
#verificar a quantidade de registros validos e invalidos
qtd_validos = df['MediaNotas'].notna().sum()
qtd_invalidos = df['MediaNotas'].isna().sum()

print("V√°lidos:", qtd_validos)
print("Inv√°lidos:", qtd_invalidos)
```

```python
#Separa√ß√£o em Dois Grupos
df_validos = df[df['MediaNotas'].notna()]
df_invalidos = df[df['MediaNotas'].isna()]
```

```python
#Quantidade de registros em cada grupo
qtd_validos = df_validos.shape[0]
qtd_invalidos = df_invalidos.shape[0]
qtd_total = df.shape[0]
```

```python
#Percentual de cada grupo
percentual_valido = (qtd_validos/qtd_total) * 100
percentual_invalido = (qtd_invalidos/qtd_total) * 100

labels = ['M√©dias Validas', 'Medias Invalidas']
valores = [percentual_valido, percentual_invalido]

plt.figure()
plt.pie(
    valores,
    labels=labels,
    autopct='%1.1f%%'
    )
plt.title('Distribui√ß√£o de M√©dias V√°lidas vs Inv√°lidas')
plt.axis('equal')
plt.show()
```

```python
perfil_municipio = df.groupby([df['MediaNotas'].notna(), 'NO_MUNICIPIO_PROVA']).size().unstack(fill_value=0)
perfil_municipio.transpose()
```

## Etapa 3 - Limpeza e Tratamento de Dados

### 1. Padroniza√ß√£o Estrutural

* Revisar e padronizar os nomes das colunas para melhor legibilidade.
* Garantir consist√™ncia sem√¢ntica entre vari√°veis similares.

```python
df.head()
```

```python
mapasCategoricos = {
    #Coluna Sexo
    'TP_SEXO': {
    'F': 'Feminino', 'M': 'Masculino'
    },
    #Coluna Estado Civil
    'TP_ESTADO_CIVIL': {
         0: 'N√£o Informado', 1: 'Solteiro(a)', 2: 'Casado(a)', 3: 'Divorciado(a)', 4: 'Viuvo(a)'
    },
    #Coluna Ra√ßa
    'TP_COR_RACA': {
        0: 'N√£o Detectado', 1: 'Branca', 2: 'Preta', 3: 'Parda', 4: 'Amarela', 5: 'Ind√≠gena', 6: 'N√£o disp√µe informa√ß√£o'
    },
    #Coluna Nacionalidade
    'TP_NACIONALIDADE': {
        0: 'N√£o Informado', 1: 'Brasileiro', 2: 'Brasileiro(a) Naturalizado', 3: 'Estrangeiro', 4: 'Brasileiro Nascido no Exterior'
    },
    #Coluna Status de Conclus√£o
    'TP_ST_CONCLUSAO': {
        1: 'J√° Conclui o Ensino Medio', 2: 'Cursando, Conclus√£o em 2021',
        3: 'Cursando, Conclus√£o ap√≥s 2021', 4: 'N√£o Concluido e n√£o Cursando o Ensino M√©dio'
    },

    'TP_ANO_CONCLUIU': {
    0: None, 1: 2020, 2: 2019, 3: 2018, 4: 2017, 5: 2016, 6: 2015, 7: 2014, 8: 2013, 9: 2012,
    10: 2011, 11: 2010, 12: 2009, 13: 2008, 14: 2007
    },
}

for coluna, mapa in mapasCategoricos.items():
    df[coluna] = df[coluna].replace(mapa)
```

```python
df['TP_ANO_CONCLUIU'] = ( df['TP_ANO_CONCLUIU']
    .replace(mapa)
    .astype('Int64')
    )

mapa_ano = {
    0: None, 1: 2020, 2: 2019, 3: 2018, 4: 2017, 5: 2016, 6: 2015, 7: 2014, 8: 2013, 9: 2012,
    10: 2011, 11: 2010, 12: 2009, 13: 2008, 14: 2007
}

df['ANO_CONCLUSAO'] = (
    df['TP_ANO_CONCLUIU']
    .replace(mapa_ano)
    .astype('Int64')
)

posAno = df.columns.get_loc('TP_ANO_CONCLUIU') + 1

df.insert(
    loc=posAno,
    column='ANO_CONCLUSAO',
    value=df.pop('ANO_CONCLUSAO')
)
```

```python
df['FAIXA_CONCLUSAO'] = None
df.loc[df['TP_ANO_CONCLUIU'] == 15, 'FAIXA_CONCLUSAO'] = 'Concluiu Antes de 2007'

pos = df.columns.get_loc('TP_ANO_CONCLUIU') + 2 

df.insert(
    loc=pos,
    column='FAIXA_CONCLUSAO',
    value=df.pop('FAIXA_CONCLUSAO')
)
```

## Etapa 4 - Engenharia de Atributos

### Planejamento Objetivo



##### **1. Cria√ß√£o de M√©trica de Desempenho Geral**



* Calcular a m√©dia geral das notas por participante.



* Consolidar o desempenho individual em uma √∫nica m√©trica num√©rica.

```python
def classificarDesempenho (nota):
    if nota < 500:
        return 'Baixo Desempenho'
    if nota < 650:
        return 'Desempenho Medio'
    else:
        return 'Alto Desempenho'
    
df['Categoria Desempenho'] = df['MediaNotas'].apply(classificarDesempenho)
```

```python
#Quantidade de pessoas em cada categoria de desempenho
df['Categoria Desempenho'].value_counts()
```

```python
#Porcentagem de pessoas em cada cat√©goria de desempenho
(
    df['Categoria Desempenho']
    .value_counts(normalize=True)
    .mul(100)
    .__round__(2)
)
```

A contagem por categoria de desempenho permite entender a distribui√ß√£o geral dos participantes

 conforme a m√©dia das notas. Essa an√°lise mostra se a maioria dos candidatos se concentra em n√≠veis
 
  baixos, m√©dios ou altos de desempenho.

##### **2. Valida√ß√£o da Categoriza√ß√£o**

```python
df.groupby('Categoria Desempenho')['MediaNotas'].describe().__round__(2)
```

A valida√ß√£o estat√≠stica da categoriza√ß√£o de desempenho demonstra uma separa√ß√£o clara e consistente entre os grupos. As faixas definidas apresentam m√©dias bem posicionadas, baixa sobreposi√ß√£o entre intervalos e dispers√£o controlada, indicando que os crit√©rios adotados s√£o adequados para segmentar o desempenho dos participantes. Al√©m disso, a distribui√ß√£o dos participantes entre as categorias reflete padr√µes esperados para avalia√ß√µes em larga escala, refor√ßando a robustez da classifica√ß√£o.

* Como os n√≠veis de desempenho se distribuem entre escolas p√∫blicas e privadas?

```python
df.groupby(['TP_ESCOLA', 'Categoria Desempenho']).size().reset_index(name='Quantidade').sort_values(by='Categoria Desempenho')
```

A contagem de participantes foi realizada utilizando o m√©todo size(), uma vez que os dados

 j√° haviam sido previamente tratados para remo√ß√£o de valores nulos nas vari√°veis de interesse,
 
  garantindo que cada linha represente um participante v√°lido para an√°lise.

```python
df_municipio = (
        df.groupby(
    ['NO_MUNICIPIO_PROVA', 'Categoria Desempenho']
        ).agg(
            Medias = ('MediaNotas', 'mean'),
            Quantidade = ('MediaNotas', 'count')
        ).reset_index().round(2)
)
```

```python
#total de participantes por municipios
total_municipios = (
    df_municipio.groupby('NO_MUNICIPIO_PROVA')['Quantidade']
    .sum()
    .reset_index(name='Total Participantes')
)
total_municipios.head().sort_values(by='Total Participantes')
```

```python
df_municipio = df_municipio.merge(
    total_municipios,
    on='NO_MUNICIPIO_PROVA',
    how='left'
)
```

Durante a agrega√ß√£o por munic√≠pio, foram identificadas colunas redundantes provenientes da etapa de jun√ß√£o dos dados.

 Essas colunas foram consolidadas, mantendo apenas o total de participantes por munic√≠pio,
 
  garantindo clareza e organiza√ß√£o do dataset final.

```python
df_municipio['Percentual'] = (
    df_municipio['Quantidade'] / df_municipio['Total Participantes'] * 100
).__round__(2)

df_municipio
```

3. Desigualdade interna (Risco Educacional)

```python
df_risco = (
    df.groupby('NO_MUNICIPIO_PROVA')
      .agg(
          Media_Geral=('MediaNotas', 'mean'),
          Desvio_Padrao=('MediaNotas', 'std'),
          Perc_Baixo_Desempenho=(
              'Categoria Desempenho',
              lambda x: (x == 'Baixo Desempenho').mean() * 100
          ),
          Total_Participantes=('MediaNotas', 'count')
      )
      .reset_index().__round__(2)
)

df_risco
```

```python
municipios_alto_risco = df_risco[
    (df_risco['Desvio_Padrao'] > df_risco['Desvio_Padrao'].quantile(0.75)) &
    (df_risco['Perc_Baixo_Desempenho'] > 50)
].sort_values('Perc_Baixo_Desempenho', ascending=False)

municipios_alto_risco
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(
    df_risco['Desvio_Padrao'],
    df_risco['Perc_Baixo_Desempenho']
)

plt.xlabel('Desvio Padr√£o das Notas')
plt.ylabel('% de Baixo Desempenho')
plt.title('Risco Educacional: Desigualdade Interna por Munic√≠pio')
plt.show()
```

Mesmo munic√≠pios com m√©dias gerais pr√≥ximas √† m√©dia estadual podem

revela munic√≠pios com maior risco educacional, caracterizados por
grande dispers√£o de resultados e concentra√ß√£o significativa de alunos
com baixo desempenho.
